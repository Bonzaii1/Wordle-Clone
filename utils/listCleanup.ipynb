## Import list

```python

# Open the text file in read mode
with open('../assets/es_5words.txt', 'r', encoding='utf-8') as file:
    # Read all lines from the file and store them in a list
    word_list = file.readlines()

# Strip newline characters from the end of each word and remove any leading/trailing whitespace
word_list = [word.strip() for word in word_list]

print(word_list)
print(len(word_list))
```

### Sort the list alphabetically

```python
word_list.sort()

print(word_list)
```

## Accents in word list

```python
accent_a = 0
accent_e = 0
accent_i = 0
accent_o = 0
accent_u = 0
for word in word_list:
    if "á" in word:
        accent_a+= 1
    elif "é" in word:
        accent_e+= 1
    elif "í" in word:
        accent_i+= 1
    elif "ó" in word:
        accent_o+= 1
    elif "ú" in word:
        accent_u+= 1

print("á: " + str(accent_a))
print("é: " + str(accent_e))
print("í: " + str(accent_i))
print("ó: " + str(accent_o))
print("ú: " + str(accent_u))
```

## Replace accented words

```python
word_list = [word.replace("á", "a").replace("é", "e").replace("í", "i").replace("ó", "o").replace("ú", "u") for word in word_list]

```

## Check if any accents remain

```python
accent_a = 0
accent_e = 0
accent_i = 0
accent_o = 0
accent_u = 0
for word in word_list:
    if "á" in word:
        accent_a+= 1
    elif "é" in word:
        accent_e+= 1
    elif "í" in word:
        accent_i+= 1
    elif "ó" in word:
        accent_o+= 1
    elif "ú" in word:
        accent_u+= 1

print("á: " + str(accent_a))
print("é: " + str(accent_e))
print("í: " + str(accent_i))
print("ó: " + str(accent_o))
print("ú: " + str(accent_u))
```

## Pass new list into new file

```python
filePath = "../assets/wordListEsOrdered.txt"

with open(filePath, "w", encoding="utf-8") as file:
    for word in word_list:
        file.write(word + ",")

print("file created successfully!")
```
